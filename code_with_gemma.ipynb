```python
# Load model directly
from transformers import AutoTokenizer, AutoModelForCausalLM
```

```python
tokenizer = AutoTokenizer.from_pretrained("google/codegemma-2b")
model = AutoModelForCausalLM.from_pretrained("google/codegemma-2b")
```

```python
prompt = "Write a python RestAPI function for calling a weather API"
```

```python
inputs = tokenizer(prompt,return_tensors="pt").to(model.device)
```

```python
inputs
```

```python
prompt_len = inputs["input_ids"].shape[-1]
```

```python
outputs = model.generate(**inputs, max_new_tokens = 150)
```

```python
outputs
```

```python
code =tokenizer.decode(outputs[0][prompt_len:])
```
